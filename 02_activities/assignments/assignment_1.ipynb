
-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
# Your answer here
def is_valid_brackets(s: str) -> bool:
    brackets = {")":"(",
                "}":"{",
                "]":"["}
    openlist = []
    ans = True
    for i in s:
        if i in brackets:
            lasti = openlist.pop() if openlist else "empty"
            if lasti != brackets[i]:
                ans = False
        else:
            openlist.append(i)
    if len(openlist) > 0:
        ans = False
    return ans
````

````python
is_valid_brackets("[]()")
````

````python
# Your answer here
# The brackets dictionary establishes the valid bracket pairs.
# Then, the openlist 'stack' is created to keep track of the
# open brackets. 
# Finally, the 'answer' is set to True before starting the for loop.
# The for loop is started to check each item in the string.
# First, the item is searched in the brackets
# dictionary. If it is there, it is a closing bracket. 
# If it is a closing bracket, the last element from the open bracket 
# list is popped from the stack.
# The last open bracket is compared to the expected open bracket
# that matches the closing bracket from the dictionary.
# If the open bracket from the dictionary doesn't match the 
# last open bracket from the stack, this means there is still
# an open bracket that has not been closed. Thus, the brackets
# are imbalanced, so the answer is changed to false.
# If there is no previous open bracket and the stack is empty, the
# popped element will be listed as "empty" which will also result in
# a non-equal check when compared to the close bracket dictionary,
# thus also resulting in an answer of false.
# Alternatively, if the item is not found in the brackets dictionary,
# it is an open bracket, so it is added to the open bracket list
# and thus to the top of the stack.
# After the entire string is run through, the open list is
# checked to see if there are any remaining items.
# Finally, if the answer was never changed to false through
# the for loop or open bracket check, it remains true
# and the brakcets are balanced.
````
